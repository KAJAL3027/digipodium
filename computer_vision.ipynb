```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = cv2.imread('parrot.jpg.jpeg')
print(a.shape)
```

```python
plt.imshow(a)
```

```python
b = cv2.cvtColor(a,cv2.COLOR_BGR2GRAY)
plt.imshow(b,cmap='gray')
```

```python
c=cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
plt.imshow(c)
```

```python
#resize by resolution
print(a.shape)
cr = cv2.resize(c,(1600,1500))
plt.imshow(cr)
```

```python
# resize by scale
cscaled = cv2.resize(c,None,fx=0.5,fy=0.5) # image reduced by 50%
plt.imshow(cscaled)
```

```python
# apply blur
cblur = cv2.GaussianBlur(c,(19,19),0)
plt.imshow(cblur)
```

```python
# apply edge detection
cgray = cv2.cvtColor(c,cv2.COLOR_RGB2BGR)
cgray = cv2.GaussianBlur(cgray, (5,5), 0)
cgary =cv2.Canny(cgray, 10, 150)
plt.imshow(cgray, cmap='gray')
```

```python
# invert image 
cinvert = cv2.bitwise_not(cgray)
plt.imshow(cinvert,cmap='gray')
```
